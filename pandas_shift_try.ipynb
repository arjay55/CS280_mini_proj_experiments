```python
import pandas as pd
```

```python
df = pd.DataFrame({"Col1": [10, 20, 15, 30, 45],

                   "Col2": [13, 23, 18, 33, 48],

                   "Col3": [17, 27, 22, 37, 52]},

                  index=pd.date_range("2020-01-01", "2020-01-05"))
```

```python
df.shift(periods=3,freq='infer')
```

```python
#number of shifts
shiftno=2
```

```python
df2=pd.DataFrame()
for _ in range(shiftno):
    df2 = pd.concat([df2,df.shift(periods=_)],axis=1)
```

```python
df2=pd.concat([df2,df2.iloc[:,:3].shift(periods=-1)],axis=1)
```
